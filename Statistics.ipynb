```python
import numpy as np
import scipy.stats as st
from scipy.special import erf
import seaborn
```

```python
norm = st.norm(loc=0.5, scale=1)

row = norm.rvs(size=50, random_state=57) # Генерация выборки из нормального распределения
row = np.sort(row)
dist = np.linspace(0, 1, 50)
row # Вариационный ряд
```

```python
f = lambda x : 0.5 * (1 + erf((x - 0.5)/np.sqrt(2))) # Функция распределения
```

```python
figure(figsize(10, 6), dpi=100)
plot(row, dist, label='Эмпирическая')
plot(row, f(row), '--', label='Теоретическая')
legend()
show()
```

```python
m = row.mean() # Выборочное среднее
m
```

```python
s = (row.std() ** 2) * 50/49 # Выборочная дисперсия
s
```

```python
t = st.t(49) # Распределение Стьюдента с 49 степенями свободы
```

```python
tau = t.ppf(0.975) # Нужный квантиль
tau
```

```python
print(m - s*tau/np.sqrt(50), m + s*tau/np.sqrt(50)) # Доверительный интервал
```
